```python
import numpy as np
import pandas as pd
import gc
from sklearn.model_selection import StratifiedKFold
import lightgbm as lgb
import catboost as cb
```

```python
from sklearn.metrics import average_precision_score

def average_precision(preds, data_vali):
    labels = data_vali.get_label()
#     preds = np.array(preds)
    return 'AP_score', average_precision_score(y_true = labels, y_score = preds), True
   

# average_precision = average_precision_score(y_test, y_score)

# print('Average precision-recall score: {0:0.2f}'.format(
#       average_precision))
```

```python
train_data = pd.read_csv('creditcard_train.csv')
test_data = pd.read_csv('creditcard_test.csv')
```

```python
test_data
```

```python
del train_data['Index']
del test_data['Index']
gc.collect()
train_y = train_data.pop('Class')
```

```python
skf = StratifiedKFold(n_splits=5, random_state=2018, shuffle=True)

xx_score = []

X = np.array(train_data)
Y = np.array(train_y)
X_test = np.array(test_data)

for k, (train_in, test_in) in enumerate(skf.split(X, Y)):
    print('train _K_ flod', k)
    X_train, X_val, y_train, y_val = X[train_in], X[test_in], Y[train_in], Y[test_in]
    
    params = {
        'boosting':'dart',
        'colsample_bytree':0.8,
        'learning_rate':0.1,
#         'metric':'auc',
        'min_child_samples':10,
        'num_leaves':95,
        'object':'binary',
        'reg_alpha':1,
        'reg_lambda':2,
        'seed':2018,
        'silent':1,
        'subsample':1.0,
        'verbose':5
    }
    train_part = lgb.Dataset(X_train,label=y_train)
    evals = lgb.Dataset(X_val,label=y_val)
    bst = lgb.train(params,train_part, 
                  num_boost_round=500, valid_sets=[train_part,evals], 
                  valid_names=['train','evals'], fobj=None,
                  feval=average_precision, feature_name='auto', 
                  categorical_feature='auto', early_stopping_rounds=20,
                  evals_result=None, verbose_eval=True, learning_rates=None, 
                  keep_training_booster=False, callbacks=None)

    
    xx_pred = bst.predict(X_val,num_iteration=bst.best_iteration)
    
    xx_score.append(average_precision_score(y_val,xx_pred,average='weighted'))
    
    y_test = bst.predict(X_test,num_iteration=bst.best_iteration)
    
    if k == 0:
        cv_pred = np.array(y_test).reshape(-1, 1)
    else:
        cv_pred = np.hstack((cv_pred, np.array(y_test).reshape(-1, 1)))
    
print(xx_score)
print np.mean(xx_score)
```
